```python

user = True 
turns = 0
board = [ ["_","_","_"],
          ["_","_","_"],
          ["_","_","_"]  ]
def print_board(board):
    for rows in board:
        for slot in rows:
            print(slot,end = " ")
        print()
    

def quit(user_input):
    if user_input.lower()=="q":
        print("thanks for playing")
        return True
    else: return False
 
def check_input(user_input):
    if is_num(user_input) == False: return False
    elif bounds(user_input) == False: return False
    else: return True
    
    

    
    
def bounds(user_input):
    user_input = int(user_input)
    if not 0<user_input<10:
        print("this number is out of bounds")
        return False
    else:
        return True
        
def is_num(user_input):
    if not user_input.isnumeric()==True:
        print("enter a valid number")
        return False
    else:
        return True
    
def is_taken(cords,board):
    row = cords[0]
    col = cords[1]
    if board[row][col] != "_":
        print("this position is already taken")
        return True
    else:
        False
    
    
def cordinates(user_input):
    rows= int(user_input/3)
    col = user_input
    if col >2: col = int(col%3)
    return (rows,col)

def add_to_board(cords,board,active_user):
    rows = cords[0]
    col  = cords[1]
    board[rows][col] = active_user
    
def current_user(user):
    if user == True:
        return "X"
    else:
        return "O"
    
def iswin(user,board):
    if check_rows(user,board) == True:return True
    if check_col(user,board) == True: return True
    if check_dia(user,board) == True : return True
    return False
    
def check_rows(user,board):
    for rows in board:
        complete_row = True
        for slot in rows:
            if slot != user:
                complete_row = False
                break
        if complete_row == True:return True
    return False

def check_col(user,board):
    for col in range(3):
        complete_col = True
        for rows in range(3):
            if board[rows][col] != user:
                complete_col = False
                break
        if complete_col == True: return True
    return False

def check_dia(user,board):
    if board[0][0] == user and board[1][1]== user and board[2][2]== user : return True
    elif board[0][2] == user and board[1][1] and board[2][0] == user : return True
    else: return False
                
while turns < 9:
    active_user = current_user(user)
    print_board(board)
    user_input = input("Enter the number through 1-9 or enter 'q' to quit: ")
    if quit(user_input)== True: break
    if check_input(user_input)== False:
        print("pls try again")
        continue
    user_input = int(user_input)-1
    cords =  cordinates(user_input)
    if is_taken(cords,board) == True:
        print("please try again")
        continue
    add_to_board(cords,board,active_user)
    if iswin(active_user,board) == True:
        print("{} won!!".format(active_user))
        break
    turns = turns +1
    if turns == 9: print("Tie")
    user = not user

    
```
